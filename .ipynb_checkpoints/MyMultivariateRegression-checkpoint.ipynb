```python
The data set contains 21613 observations with 21 features
Descriptions and names of the columns (features) are given below.

            1. id
            2. date: Date house was sold
            3. price: Price of the sold house
            4. bedrooms: Number of Bedrooms
            5. bathrooms: Number of bathrooms
            6. sqft_living: Square footage of the living space
            7. sqrt_lot: Square footage of the lot
            8. floors: Total floors in the house
            9. waterfront: Whether the house is on a waterfront(1: yes, 0: no)
            10. view: special view?
            11. condition: Condition of the house
            12. grade: unknown
            13. sqft_above: Square footage of house apart from basement
            14. sqft_basement: Square footage of the basement
            15. yr_built: Built year
            16. yr_renovated: Year when the house was renovated
            17. zipcode: zipcode of the house
            18. lat: Latitude coordinate
            19. long Longitude coordinate
            20. sqft_living15: Living room area in 2015(implies some renovations)
            21. sqrt_lot15: Lot area in 2015(implies some renovations)
                
The prices of the house  is our target variable 
and the remaining are the feature variables based on which we will predict the value of a house.
```

```python
#let check the for the correlation between our target variable which is price and all other variables.
#this helps in determing  the  input variables  for our model

correlation_matrix=data.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

From the heatmap above we can see that there is strong correlation between 
grade (0.67), sqft_living(0.7), and our target variable (price).


```python
#Can we visualize the correlation?

plt.figure(figsize=(20,5))
features=[ 'grade','sqft_living']
target=data['price']
for i, col in enumerate(features):
    plt.subplot(1, len(features), i+1)
    x=data[col]
    y=target
    plt.scatter(x,y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('price')
```

Prepare the data for training.
since we observed that grade and sqft_living are the variables  with higher correlation with our target variable.
they will serve as our feature variable and price as target variable.

```python
X=pd.DataFrame(np.c_[data['grade'], data['sqft_living']], columns=['grade', 'sqft_living'])
Y=data['price']
```

```python
Split data into train and test set
```

```python
#import the libraries
import sklearn
from sklearn import datasets, linear_model

from sklearn.cross_validation import train_test_split



X_train, X_test, Y_train, Y_test= sklearn.cross_validation.train_test_split(
X, Y, test_size=0.30, random_state=5)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

train the model using sklearn linearRgression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_model=LinearRegression()
lin_model.fit(X_train,Y_train)
```

```python
#model evaluation for training set

y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

```

```python
# model evaluation for testing set

y_test_predict = lin_model.predict(X_test)
# root mean square error of the model
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))

# r-squared score of the model
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
# plotting the y_test vs y_pred

plt.scatter(Y_test, y_test_predict)
plt.show()
```
